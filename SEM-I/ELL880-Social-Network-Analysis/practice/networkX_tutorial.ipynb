```python
# create a directed random graph on 100 nodes and 1k edges
G2 = snap.GenRndGnm(snap.TNGraph, 100, 1000)
# traverse the nodes
for NI in G2.Nodes():
    print("node id %d with out-degree %d and in-degree %d" % (
        NI.GetId(), NI.GetOutDeg(), NI.GetInDeg()))
# traverse the edges
for EI in G2.Edges():
    print("edge (%d, %d)" % (EI.GetSrcNId(), EI.GetDstNId()))
# traverse the edges by nodes
for NI in G2.Nodes():
    for Id in NI.GetOutEdges():
        print("edge (%d %d)" % (NI.GetId(), Id))
```

### Maximum flow problem

```python
G7 = nx.DiGraph()
```

```python
G7.add_nodes_from(["S","A","B","C","E"])
```

```python
G7.add_edges_from([
    ("S","A",{"capacity":12}),
    ("S","C",{"capacity":5}),
    ("S","B",{"capacity":8}),
    ("A","C",{"capacity":9}),
    ("A","B",{"capacity":10}),
    ("B","E",{"capacity":5}),
    ("C","B",{"capacity":7}),
    ("C","E",{"capacity":16}),
])
```

```python
pos = {
    "S":(1,5),
    "A":(4.6,2.6),
    "B":(3,8),
    "C":(7,3),
    "E":(8,9),
}
pos_attribute = {
    "S":(0.3,5),
    "A":(3.9,2.4),
    "B":(2.2,8),
    "C":(7.7,3.2),
    "E":(8.7,9.2),
}
nx.draw(G7, pos = pos,
        with_labels = True, node_color = "orange",
        node_size = 600, font_color = "black",
        font_size = 15, font_family = "Sans serif",
        # font_weight = "bold"
        width = 3, edgecolors="black")
nx.draw_networkx_labels(G7,pos = pos_attribute,
                        font_color = "black",
                        font_size = 15, font_family = "Sans serif",)
plt.margins(.3)
plt.title("GRAPH 6", color = "black", size = 20)
plt.show()
```

```python
from networkx.algorithms.flow import shortest_augmenting_path
```

```python
nx.maximum_flow(G7,"S","E",flow_func = shortest_augmenting_path)
```

```python
flow_value, flow_dict = nx.maximum_flow(G7,"S","E",flow_func = shortest_augmenting_path)
```
